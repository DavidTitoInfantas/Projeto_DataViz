```python
df_data_mean_price = df_clean.groupby(['date_of_reference','brand']).avg_price_brl.mean().round(2).unstack(fill_value=0).reset_index()
```

```python
## Dataset com preço medio por marca
df_data_mean_price.head()
```

```python
columns_brand = list(df_data_mean_price.columns)
brand_removes = ['date_of_reference','Buggy','CAB Motors'] ## Removemos marcas que não tem dados em mais do 50% do periodo
columns_brand = list(filter(lambda x: x not in brand_removes,columns_brand))
#columns_brand.remove('date_of_reference','Buggy','CAB Motors')
```

```python
columns_variacao = []
for i in columns_brand:
    variacao_brand = f'variacao_{i}_%'
    df_data_mean_price[variacao_brand] = round(((df_data_mean_price[i].diff()/df_data_mean_price[i]))*100,2)
    columns_variacao.append(variacao_brand)
#columns_variacao.remove('variacao_date_of_reference')
```

```python
df_variacao = df_data_mean_price[columns_variacao]
```

```python
df_variacao['Mean_ABS_variacao_bands'] = df_variacao[columns_variacao].abs().sum(axis=1)
df_variacao['Mean_variacao_bands'] = df_variacao[columns_variacao].sum(axis=1)
```

```python
def plotar(title,xlabel,ylabel, x, y,df):
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    ax = sns.lineplot(x=x, y=y, data = df)
    ##as1= sns.lineplot(x=x, y=y1, data = df)
    ax.figure.set_size_inches(10,5)
    ax.set_title(title,loc = 'left', fontsize=18)
    ax.set_xlabel(xlabel,fontsize=14)
    ax.set_ylabel(ylabel,fontsize=14)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def scatter_text(x, y, text_column, data, title, xlabel, ylabel):
    """Scatter plot with country codes on the x y coordinates
       Based on this answer: https://stackoverflow.com/a/54789170/2641825"""
    # Create the scatter plot
    p1 = sns.scatterplot(x, y, data=data, size = 8, legend=False)
    # Add text besides each point
    for line in range(0,data.shape[0]):
         p1.text(data[x][line]+0.01, data[y][line], 
                 data[text_column][line], horizontalalignment='left', 
                 size='medium', color='black', weight='semibold')
    # Set title and axis labels
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    return p1
```

```python
teste = df_analise_2.groupby(['year_of_reference','fipe_code', 'brand', 'fuel', 'gear', 'engine_size']
                            ).avg_price_brl.mean()
```

```python
teste = pd.DataFrame(teste).reset_index()
teste.head()
```

```python
#Deletando e arrumando os dados
teste = teste[teste['fuel'] != 'Alcohol']
teste = teste[(teste['fipe_code'] != '040001-7') & (teste['brand'] != 'Buggy')]
teste = teste[(teste['fipe_code'] != '035055-9') & (teste['engine_size'] != '2.9')]
teste = teste[(teste['fipe_code'] != '017006-2') & (teste['gear'] != 'manual')]
```

```python
verificacao = teste.groupby('fipe_code').year_of_reference.count()
verificacao = pd.DataFrame(verificacao).reset_index().sort_values(by = 'year_of_reference')
verificacao = verificacao[verificacao['year_of_reference']<2]['fipe_code']
teste = teste[~teste['fipe_code'].isin(list(verificacao))]
```

```python
teste = teste.sort_values(by='fipe_code')
teste.head(5)
```

```python
teste = teste.pivot_table(index=['fipe_code', 'brand', 'fuel', 'gear', 'engine_size'],
                          columns='year_of_reference', 
                          values = 'avg_price_brl').reset_index()
teste.rename(columns={2021:'year_2021',2022:'year_2022'},inplace=True)
```

```python
teste['taxa_variacao'] = round((teste['year_2022'] - teste['year_2021'])/teste['year_2021']*100,2)
teste['year_2021'] = teste['year_2021'].round(2)
teste['year_2022'] = teste['year_2022'].round(2)
```

```python
teste = teste.merge(df_analise_2[['model','fipe_code','year_model']],
                    how='left',left_on = 'fipe_code',right_on='fipe_code',)
teste= teste.groupby(['fipe_code', 'brand', 'fuel', 
                      'gear', 'engine_size', 'year_2021', 
                      'year_2022', 'taxa_variacao'])[['model','year_model']].first().sort_values(
                        by='taxa_variacao',ascending=False).reset_index()
```

```python
#segmentação de acordo a o que procuramos
grupo_1 =teste[(teste['year_2022'] <= 35000) & 
(teste['taxa_variacao'] >= 15) ]
##(teste['engine_size'] >= 15) &
```

```python
grupo_1.head()
```

```python
##criar score com pesos especificos, considerando os seguintes fatores:
# Preço do carro
# Taxa de variação de preço
# Ano de fabricação do carro

grupo_1['Score']  = (((grupo_1['year_2022'] / grupo_1['year_2022'].max()-1).abs())*0.2) + (
                    (grupo_1['taxa_variacao'] / grupo_1['taxa_variacao'].max())*0.2) + (
                    (((2022 - grupo_1['year_model']) / (2022 - grupo_1['year_model']).max()-1).abs())*0.6)
```

```python
grupo_1 = grupo_1.sort_values(by='Score',ascending = False).reset_index(drop=True)
```

```python
grupo_1.head(30) 
```

```python
def plot_comparacao(x, y1, y2, y3, dataset, titulo):
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    plt.figure(figsize=(16,12))
    ax=plt.subplot(3,1,1)
    ax.set_title(titulo,fontsize=18,loc='left')
    sns.lineplot(x = x,y=y1,data=dataset)
    plt.subplot(3,1,2)
    sns.lineplot(x = x,y=y2,data=dataset)
    plt.subplot(3,1,3)
    sns.lineplot(x = x, y=y3,data = dataset)
    ax=ax
```
